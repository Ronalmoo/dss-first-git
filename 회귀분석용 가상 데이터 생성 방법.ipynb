```python
def make_regression2(n_samples, bias, noise, random_state):
    np.random.seed(random_state)
    noise = scipy.stats.norm(0, noise).rvs(n_samples)
    X = np.random.randn(1, n_samples)
    w = np.random.randn()
    y = X * w + bias + noise
    return X, w, y
```
